```python
#importar las librerías:
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#para importar se requiere utilizar el comando
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Cesde/autores.csv')
```

# Comandos 2:
 Generar gráfico de dispersión

```python
nuevo_data = data.loc[data['Universidad']=='Universidad de Bolivia']
nuevo_data.head(5)
```

```python
plt.scatter(nuevo_data['Edad'],nuevo_data['Publicaciones'])
plt.title('Grafico de dispersión Universidad de Bolivia')
plt.xlabel('Edad')
plt.ylabel('Publicaciones')
plt.show()
```

#Comandos 3:
Uso de un gráfico de pie.

```python
circular = data.groupby('Nacionalidad')['Nacionalidad'].count()

x = circular.index
y = circular.values

plt.pie(y,
        autopct='%1.1f%%',
        labels=x)
plt.title('Proporción de autores por nacionalidad')
plt.show()
```

#Comandos 4:

Uso de un gráfico de caja.

```python
'''con esta línea quiero visualizar la cantidad de datos que tiene cada pais
para determinar si vale la pena construir un caja y bigotes comparativo'''

data.groupby('Nacionalidad')['Nacionalidad'].count()
```

```python
sns.boxplot(
    data=data,
    y='Publicaciones',
    hue='Nacionalidad'
)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#Comandos 5:
Uso de gráficos comparativos.

```python
#countplot

sns.countplot(
    data=data,
    x='Nacionalidad',
    hue='Sexo'

)

plt.xticks(rotation=45, ha='right')
plt.show()
```


#Comandos 6:

Creación de gráficos usando la librería de seaborn.

```python
sns.stripplot( data=data, x='Universidad', y='Edad', hue='Sexo', legend=False, dodge=True, color='black')

sns.boxplot( data=data, x='Universidad', y='Edad', hue='Sexo', legend=True)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```


# Comandos 7:

Uso de histograma + box plot

```python
#Histograma:

sns.histplot(
    data=data,
    x='Publicaciones'
)

# Mostrar el gráfico
plt.show()
```
